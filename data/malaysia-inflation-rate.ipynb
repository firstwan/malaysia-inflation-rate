```python
import xml.etree.ElementTree as ET
tree = ET.parse(r"./WEO_PUB_APR2022.xml")
root = tree.getroot()

import csv
f = open('./weo_pub_converted_file.csv', 'w', encoding='utf_8', newline='')
writer = csv.writer(f)

header = ['unit', 'concept', 'ref_area', 'freq', 'last_actual_date', 'scale',
       'notes', 'time_period', 'obs']

writer.writerow(header)
all_list = []
count = 0

for series in range(len(root)):
    base_value = []
    series_attb = root[series].attrib
    for item in series_attb:
        base_value.append(series_attb[item])

    for obs in range(len(root[series])):
        year_value = []
        year_value.append(int(root[series][obs].get("TIME_PERIOD")))

        obs_value = root[series][obs].get("OBS_VALUE")
        year_value.append(float(obs_value) if obs_value.isdigit() else None)

        try:
            writer.writerow(base_value + year_value)
        except:
            print(base_value + year_value)
            raise Exception("faile")
            
        count += 1

        if count > 10000:
            count = 0
            f.close()
            f = open('./weo_pub_converted_file.csv', 'a+', encoding='utf_8', newline='')
            writer = csv.writer(f)
            
        # all_list.append(base_value + year_value)

# writer.writerows(all_list)

f.close()
```
